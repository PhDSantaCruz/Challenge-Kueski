En este cuaderno calculamos cinco características para el conjunto de datos de **riesgo crediticio**.
Cada fila en el conjunto de datos consiste en el crédito que tomó un usuario en una fecha determinada.

Estas características se definen aproximadamente de la siguiente manera:

**nb_previous_loans:** número de préstamos otorgados a un usuario determinado, antes del préstamo actual.

**avg_amount_loans_previous:** monto promedio de préstamos otorgados a un usuario, antes de la calificación actual.

**edad:** edad del usuario en años.

**años_en_el_trabajo:** años que el usuario ha estado en el empleo.

**bandera_coche_propio:** bandera que indica si el usuario tiene coche propio.

Tenemos el siguiente problema: la función `avg_amount_loans_previous` tarda demasiado en calcularse para todas las filas del conjunto de datos (al menos en la forma en que se implementa). 



```python
import pandas as pd
```

```python
df = pd.read_csv('dataset_credit_risk.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.sort_values(by=["id", "loan_date"])
df = df.reset_index(drop=True)
df["loan_date"] = pd.to_datetime(df.loan_date)
df
```

#### Feature nb_previous_loans

```python
# Agrupamos por  'id'  
df_grouped = df.groupby("id")
#Asignar número de creditos anteriores a cada nuevo crédito
df["nb_previous_loans"] = df_grouped["loan_date"].rank(method="first") - 1
```

```python
df[['id',"nb_previous_loans","loan_date"]].head(20)
```

#### Feature avg_amount_loans_previous

```python
#Asignar el monto promedio de creditos anteriores
def avg_amount_loans_prev(df):
    avg = pd.Series(index=df.index)
    for i in df.index:
        df_aux = df.loc[df.loan_date < df.loan_date.loc[i], :]
        avg.at[i] = df_aux.loan_amount.mean()
    return avg
```

```python
avg_amount_loans_previous = pd.Series()

# the following cycle is the one that takes forever if we try to compute it for the whole dataset
# el siguiente ciclo es el que tarda una eternidad si intentamos calcularlo para todo el conjunto de datos 

for user in df.id.unique():
    df_user = df.loc[df.id == user, :]
    avg_amount_loans_previous = avg_amount_loans_previous.append(avg_amount_loans_prev(df_user))
```

```python
df["avg_amount_loans_previous"] = avg_amount_loans_previous
```

#### Feature age

```python
from datetime import datetime, date
```

```python
df['birthday'] = pd.to_datetime(df['birthday'], errors='coerce')
```

```python
df['age'] = (pd.to_datetime('today').normalize() - df['birthday']).dt.days // 365
```

#### Feature years_on_the_job

```python
df['job_start_date'] = pd.to_datetime(df['job_start_date'], errors='coerce')
```

```python
df['years_on_the_job'] = (pd.to_datetime('today').normalize() - df['job_start_date']).dt.days // 365
```

#### Feature flag_own_car

```python
df['flag_own_car'] = df.flag_own_car.apply(lambda x : 0 if x == 'N' else 1)
```

## Save dataset for model training

```python
df = df[['id', 'age', 'years_on_the_job', 'nb_previous_loans', 'avg_amount_loans_previous', 'flag_own_car', 'status']]
```
